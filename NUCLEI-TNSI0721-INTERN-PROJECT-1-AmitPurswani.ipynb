```python
## Imputing Null Values for Item_Weight
## Since its Numerical and Continuous we will replace Null values with its mean value
data['Item_Weight'].median()
```

```python
## Imputing Null values in 'Item_Weight'
data['Item_Weight']=data['Item_Weight'].fillna(data['Item_Weight'].median())
```

```python
## Check if Null Values have been imputed
data['Item_Weight'].isnull().sum()
```

```python
##Check and Treat Null Values of Outlet_Size
## As it is Categorical we will impute Null values with its Mode value
data['Outlet_Size'].mode()
```

```python
## Impute its Null Values
data['Outlet_Size']=data['Outlet_Size'].fillna('Medium')
```

```python
data['Outlet_Size'].isnull().sum()
```

```python
##Check overall Null Values
data.isnull().sum()
```

```python
## Although Item_Visibility does not have Null Value but it can't have Zero value for certain items,
##as items will have some visibility however small it might be
## so replacing with its median value
data['Item_Visibility'].median()
```

```python
data['Item_Visibility']=data['Item_Visibility'].replace(to_replace=0, value=0.054020)
```

```python
data['Item_Visibility'].value_counts()
```

```python
## Replacing 'LF' and 'low fat' with 'Low Fat'
data['Item_Fat_Content']=data['Item_Fat_Content'].replace(to_replace='LF', value='Low Fat')
data['Item_Fat_Content']=data['Item_Fat_Content'].replace(to_replace='low fat', value='Low Fat')
## Replacing 'reg' with 'Regular'
data['Item_Fat_Content']=data['Item_Fat_Content'].replace(to_replace='reg', value='Regular')
```

```python
data['Item_Fat_Content'].value_counts()
```

```python
## Outlet Count
outlet_count = data['Outlet_Identifier'].value_counts()
```

```python
## Item count
item_count = data['Item_Identifier'].value_counts()
```

```python
item_count
```

```python
data['Outlet_Years']=2013 - data['Outlet_Establishment_Year']
```

```python
data.head()
```

```python
## making a new column Item_Identifier_New for categorizing in three broad categories
data['Item_Identifier_New'] = data['Item_Identifier'].apply(lambda x : x[0:2])
```

```python
data.head()
```

```python
## expanding the category to appropriate name identifier
data['Item_Identifier_New']=data['Item_Identifier_New'].map({'FD':'Food','NC':'Non-Consumable','DR':'Drinks'})
```

```python
## Categorical Variables are 'Item_Fat_Content', 'Outlet_Size', 'Outlet_Location_Type', 'Outlet_Type', Item_Identifier_New
```

```python
##Check Fat content where Items are Non-Consumable
data[data['Item_Identifier_New'] =='Non-Consumable']
```

```python
##Replace the fat content of Non-Consumable Items as Non-Edible
data.loc[data['Item_Identifier_New'] =='Non-Consumable', 'Item_Fat_Content'] = "Non-Edible"
```

```python
##Check again
data[data['Item_Identifier_New'] =='Non-Consumable']
```

```python
#One Hot Coding:
data = pd.get_dummies(data, columns=['Item_Fat_Content', 'Outlet_Size', 'Outlet_Location_Type', 'Outlet_Type', 'Outlet_Identifier', 'Item_Identifier_New'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
y = train1['Item_Outlet_Sales']
X = train1.drop(columns=['Item_Outlet_Sales'],axis=1)
```

## Linear Regression Model

```python
## Referred https://www.analyticsvidhya.com/blog/2016/02/bigmart-sales-solution-top-20/
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train, y_train)
```

```python
y_pred=lm.predict(X_test)
```

```python
from sklearn import metrics
```

## Linear Regression Metrics

```python
print('MAE', metrics.mean_absolute_error(y_test,y_pred))
print('MSE', metrics.mean_squared_error(y_test,y_pred))
print('RMSE', np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
print('R2 score', metrics.r2_score(y_test,y_pred))
```

## Decision Trees Model

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dtree = DecisionTreeRegressor()
```

```python
dtree.fit(X_train, y_train)
```

```python
y_pred = dtree.predict(X_test)
```

## Decision Tree Metrics

```python
print('MAE', metrics.mean_absolute_error(y_test,y_pred))
print('MSE', metrics.mean_squared_error(y_test,y_pred))
print('RMSE', np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
print('R2 score', metrics.r2_score(y_test,y_pred))
```

## Random Forest Model

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfc = RandomForestRegressor(n_estimators=300)
```

```python
rfc.fit(X_train,y_train)
```

```python
y_pred = rfc.predict(X_test)
```

## Random Forest Metrics

```python
print('MAE', metrics.mean_absolute_error(y_test,y_pred))
print('MSE', metrics.mean_squared_error(y_test,y_pred))
print('RMSE', np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
print('R2 score', metrics.r2_score(y_test,y_pred))
```

## XGBoost Model

```python
## Referring https://machinelearningmastery.com/xgboost-for-regression/
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
y_pred = xgb.predict(X_test)
```

## XGBoost Metrics

```python
print('MAE', metrics.mean_absolute_error(y_test,y_pred))
print('MSE', metrics.mean_squared_error(y_test,y_pred))
print('RMSE', np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
print('R2 score', metrics.r2_score(y_test,y_pred))
```
